Work out the order that transactions occur for each account
Hint: where multiple transactions happen on the same day, assume the highest value transactions happen first

```python
total = pd.concat([incoming,outgoing])
```

```python
total
```

```python
data_1 = pd.concat([df,total])
```

```python
data_1.head(5)
```

```python
data_1.sort_values(by=['Account', 'Balance Date', 'Value'], ascending=[True, True, False], inplace=True)
```

```python
data_1
```

Use a running sum to calculate the Balance for each account on each dayou

```python
data_1 = data_1[['Account', 'Balance Date','Value']]
```

```python
data_1['Balance'] = data_1.groupby('Account')['Value'].cumsum()
```

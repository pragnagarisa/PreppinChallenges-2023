```python
import pandas as pd
import numpy as np
```

```python
data_dict = pd.read_excel("Student_Grades_input.xlsx",sheet_name = None)
```

```python
data_dict.keys()
```

```python
df = pd.DataFrame()
```

```python
for sheet in data_dict:
    if sheet != 'Headmaster Notes':
        df = pd.concat([df,pd.read_excel("Student_Grades_input.xlsx",sheet_name = sheet)
                             .assign(Term = sheet)])
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'],format = 'mixed')
df['Date'] = df['Date'].dt.strftime('%d/%m/%Y')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'].info()
```

```python
#Another Dataset

df1 = pd.read_csv("Student_LookupTable.csv")
```

```python
df1.head(5)
```

```python
df1.info()
```

```python
#Merging


data = df1.merge(df,how = 'inner',on = 'id')
```

```python
data.head(5)
```

```python
data.shape
```

```python
data.columns
```

```python
#GPA

data['GPA'] = data.loc[:,['Subject_1_Grade','Subject_2_Grade', 'Subject_3_Grade', 'Subject_4_Grade']].mean(axis = 1)
```

```python
data
```

```python
# 3 Term GPA Moving


data['3 Term GPA Moving'] = data['GPA'].rolling(window=3,min_periods=3).mean().round(2)
```

```python
data
```

```python
#Sorting

data.sort_values(by = ['first_name','last_name','Date'],ascending = [False,False,True],inplace = True)
```

```python
data.head(5)
```

```python
#Full name

data['Student_Name'] = data['first_name']+' '+data['last_name']
```

```python
data.head(5)
```

```python
#Re-index columns

cols = ['id','Student_Name','Date','Term','GPA','3 Term GPA Moving']
data = data.reindex(columns = cols)
```

```python
data.head(5)
```

```python
#Random Checking

data[data['Student_Name'] == 'Sadie Ashling']
```

```python
#Reindex

data.index = np.arange(1,len(data)+1)
```

```python
data.head(5)
```

```python
data['Rank'] = data.index.argsort().argsort() + 1
```

```python
data.head(5)
```

```python
cols = ['Rank','id','Student_Name','Date','Term','GPA','3 Term GPA Moving']

data = data.reindex(columns = cols)
```

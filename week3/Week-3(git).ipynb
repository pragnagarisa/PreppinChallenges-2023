```python
1. For the transactions file:
   (1) Filter the transactions to just look at DSB 
   (2) These will be transactions that contain DSB in the Transaction Code field
   (3) Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values
   (4) Change the date to be the quarter 
   (5) Sum the transaction values for each quarter and for each Type of Transaction (Online or In-Person) 
```

```python
1.1. Filter the transactions to just look at DSB 
```

```python
data_1['Transaction Code'] = data_1['Transaction Code'].str.split("-",expand = True)[0]
```

```python
data_dsb = data_1[data_1['Transaction Code'] == 'DSB']
```

```python
1.2. These will be transactions that contain DSB in the Transaction Code field
```

```python
data_dsb
```

```python
1.3. Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values
```

```python
values = {1:"Online",2:"In-Person"}
```

```python
data_dsb['Online or In-Person'] = data_dsb['Online or In-Person'].replace(values)
```

```python
data_dsb
```

```python
1.4. Change the date to be the quarter
```

```python
data_dsb['Transaction Date'] = pd.to_datetime(data_dsb['Transaction Date'])
data_dsb['Transaction Date'] = data_dsb['Transaction Date'].dt.quarter
```

```python
data_dsb
```

```python
1.5. Sum the transaction values for each quarter and for each Type of Transaction (Online or In-Person) 
```

```python
final_dsb = data_dsb.groupby(['Online or In-Person','Transaction Date'])['Value'].sum()
```

```python
final_dsb = final_dsb.reset_index(name = "Value")
```

```python
final_dsb.sort_values(by = 'Transaction Date',inplace = True)
```

```python
final_dsb.rename(columns = {'Transaction Date':'Quarter'},inplace = True)
```

```python
final_dsb
```

```python
2. For the targets file:
   1. Pivot the quarterly targets so we have a row for each Type of Transaction and each Quarter 
   2. Rename the fields
   3. Remove the 'Q' from the quarter field and make the data type numeric 
```

```python
data_2 = pd.read_csv("Targets.csv")
```

```python
data_2
```

```python
data_2.info()
```

```python
2.1. Pivot the quarterly targets so we have a row for each Type of Transaction and each Quarter 
```

```python
2.2. Rename the fields
```

```python
pivot_melt = data_2.melt(id_vars=["Online or In-Person"] ,var_name = 'Quarter', value_name = "Quarter_targets")
```

```python
pivot_melt
```

```python
2.3. Remove the 'Q' from the quarter field and make the data type numeric
```

```python
pivot_melt['Quarter'] = pivot_melt['Quarter'].str.replace("Q","")
```

```python
pivot_melt
```

```python
pivot_melt['Quarter'] = pivot_melt['Quarter'].astype(int)
```

```python
pivot_melt.info()
```

```python
3. Join the two datasets together    
   You may need more than one join clause!
```

```python
final = final_dsb.merge(pivot_melt , how= 'inner', on = ['Online or In-Person','Quarter'])
```

```python
final
```

```python
order = ['Online or In-Person','Quarter','Value','Quarter_targets']
```

```python
final = final[order]
```

```python
final
```

```python
final['Variance to Targets'] = final['Value'] - final['Quarter_targets']
```

```python
final
```

```python
final.rename(columns = {'Quarter_targets': 'Quarterly Targets'},inplace = True)
```

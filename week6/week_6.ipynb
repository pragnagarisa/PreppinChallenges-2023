```python
import pandas as pd
```

```python
data = pd.read_csv("DSB Customer Survery.csv")
```

```python
data.head(5)
```

```python
#Exclude the Overall Ratings, these were incorrectly calculated by the system
```

```python
data.drop(columns = ['Mobile App - Overall Rating','Online Interface - Overall Rating'],axis = 1,inplace = True)
```

```python
#Reshape the data so we have 5 rows for each customer, with responses for the Mobile App and Online Interface being in separate fields on the same row
```

```python
tab = data.melt(id_vars = 'Customer ID',var_name = "Online/Mobile" ,value_name = "Values")
```

```python
tab
```

```python
tab['Online-Mobile'] = tab['Online/Mobile'].str.split("-",expand = True)[0]
```

```python
tab
```

```python
Clean the question categories so they don't have the platform in from of them
     e.g. Mobile App - Ease of Use should be simply Ease of Use
```

```python
tab['Question'] = tab['Online/Mobile'].str.split("-",expand = True)[1]
```

```python
tab
```

```python
tab.drop(columns = 'Online/Mobile',axis = 1 , inplace = True)
```

```python
tab
```

```python
tab = tab.reindex(columns = ['Customer ID','Question','Online-Mobile','Values'])
```

```python
tab
```

```python
Calculate the Average Ratings for each platform for each customer 
```

```python
data_1 = tab.groupby(['Customer ID','Online-Mobile'])['Values'].mean().reset_index()
```

```python
data_1
```

```python
Calculate the difference in Average Rating between Mobile App and Online Interface for each customer
```

```python
final = data_1.pivot_table(index = 'Customer ID',columns = 'Online-Mobile',values = 'Values')
```

```python
final
```

```python
final.info()
```

```python
final.columns
```

```python
final['Difference'] = final['Mobile App '] - final['Online Interface ']
```

```python
final
```

```python
Catergorise customers as being:
 - Mobile App Superfans if the difference is greater than or equal to 2 in the Mobile App's favour
 - Mobile App Fans if difference >= 1
 - Online Interface Fan
 - Online Interface Superfan
 - Neutral if difference is between 0 and 1
```

```python
import numpy as np

final['Preference'] = np.where(final['Difference'] <= -2, 'Online Interface Superfan',
                      np.where(final['Difference'] <= -1, 'Online Interface Fan',
                      np.where(final['Difference'] < 1, 'Neutral',
                      np.where(final['Difference'] < 2, 'Mobile App Fan',
                              'Mobile App Superfan'))))
```

```python
final
```

```python
total = final['Preference'].value_counts()
```

```python
output = (total/len(final))*100
```

```python
output
```

```python
output.to_csv("Week_6_output.csv")
```

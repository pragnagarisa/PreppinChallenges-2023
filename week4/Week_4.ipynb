```python
import pandas as pd
```

```python
data_dict = pd.read_excel("New Customers.xlsx",sheet_name = None)
```

```python
data_dict.keys()
```

```python
data_dict.get('January')
```

```python
fields = {'Demographiic' : 'Demographic','Demagraphic' : 'Demographic'}
```

```python
df = pd.DataFrame()
```

```python
for sheet in data_dict.keys():
    df = pd.concat([pd.read_excel("New Customers.xlsx",sheet_name = sheet)
                                 .assign(Month = sheet)
                                 .rename(columns = fields),df])
```

```python
df
```

```python
nums = {'January':'1', 'February':'2', 'March':'3', 'April':'4', 'May':'5', 'June':'6', 'July':'7', 'August':'8',
        'September':'9', 'October':'10','November':'11', 'December':'12'}
```

```python
df['Month'] = df['Month'].replace(nums)
```

```python
df
```

```python
df['Joining Date'] = df['Month'].astype(str) + '/' + df['Joining Day'].astype(str) + '/2023'
```

```python
df
```

```python
final = df.pivot(index = ['ID','Joining Date'],columns = 'Demographic' ,values = 'Value')
```

```python
final.to_csv('Week4_output.csv')
```

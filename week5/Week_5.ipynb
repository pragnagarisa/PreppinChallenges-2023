```python
import pandas as pd
```

```python
data = pd.read_csv("PD 2023 Wk 1 Input.csv")
```

```python
data.head(5)
```

```python
data.info()
```

```python
1. Create the bank code by splitting out off the letters from the Transaction code, call this field 'Bank'
```

```python
data['Transaction Code'] = data['Transaction Code'].str.split("-",expand = True)[0]
```

```python
data
```

```python
2. Change transaction date to the just be the month of the transaction
```

```python
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'])
data['Transaction Date'] = data['Transaction Date'].dt.month_name()
```

```python
data
```

```python
3. Total up the transaction values so you have one row for each bank and month combination
```

```python
d = data.groupby(['Transaction Code','Transaction Date'])['Value'].sum()
data_1 = d.reset_index(name = 'values')
```

```python
data_1
```

```python
4. Rank each bank for their value of transactions each month against the other banks. 1st is the highest value
   of transactions, 3rd the lowest. 
```

```python
data_1['Rank'] = data_1.groupby(['Transaction Date'])['values'].rank(ascending = False)
```

```python
data_1
```

```python
5. Without losing all of the other data fields, find:
   (1) The average rank a bank has across all of the months, call this field 'Avg Rank per Bank'
   (2) The average transaction value per rank, call this field 'Avg Transaction Value per Rank'
```

```python
#'Avg Transaction Value per Rank'

data_1.groupby(['Rank'])['values'].mean().reset_index()
```

```python
for index,row in data_1.iterrows():
     if row['Rank'] == 1.0 :
        data_1.at[index,'Avg Transaction Value per Rank'] = 66967.750000
     elif row ['Rank'] == 2.0:
        data_1.at[index,'Avg Transaction Value per Rank'] = 48633.666667
     else:
        data_1.at[index,'Avg Transaction Value per Rank'] = 34620.833333
          
    
```

```python
data_1
```

```python
#'Avg Rank per Bank'

data_1.groupby('Transaction Code')['Rank'].mean()
```

```python
for index,row in data_1.iterrows():
    if row['Transaction Code'] == 'DS':
        data_1.at[index,'Avg Rank per Bank'] = 1.916667
    elif row['Transaction Code'] == 'DSB':
        data_1.at[index,'Avg Rank per Bank'] = 2.333333
    else:
        data_1.at[index,'Avg Rank per Bank'] = 1.750000
```

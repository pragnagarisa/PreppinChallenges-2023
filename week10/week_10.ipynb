```python
data.drop(columns = 'Unnamed: 0',inplace = True)
```

```python
data['Balance Date'] = pd.to_datetime(data['Balance Date'])
```

Aggregate the data so we have a single balance for each day already in the dataset, for each account

```python
df = data.groupby(['Account','Balance Date'])[['Balance','Transaction Value']].sum().reset_index()
```

```python
df.head(5)
```

Scaffold the data so each account has a row between 31st Jan and 14th Feb (hint)

```python
start_date = pd.to_datetime('2023-01-31')
end_date = pd.to_datetime('2023-02-14')
```

```python
all_days = pd.DataFrame(pd.date_range(start_date,end_date,freq = '1D'),columns = ['Balance Date'])
```

```python
all_days['Key'] = 1
df['Key'] = 1
```

```python
df_1 = all_days.merge(df['Account'],how = 'cross')
```

```python
df_1.head(5)
```

```python
df.head(5)
```

```python
total = df_1.merge(df,how = 'left',on = ['Account','Key','Balance Date'])
```

```python
total
```

```python
total = total.drop_duplicates()
```

```python
final = total.merge(data,how = 'left',on = ['Account','Balance Date','Balance'])
```

```python
cols = ['Account','Balance Date','Transaction Value_x','Balance']
```

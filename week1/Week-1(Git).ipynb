```python
import pandas as pd
```

```python
data = pd.read_csv("PD 2023 Wk 1 Input.csv")
```

```python
Observation:
    1. Have to change "Transaction Date" format
```

```python
data["Transaction Date"] = pd.to_datetime(data["Transaction Date"])
data["Transaction Date"] = data["Transaction Date"].dt.date
```

```python
data
```

```python
1. Split the Transaction Code to extract the letters at the start of the transaction code. 
  These identify the bank who processes the transaction . Rename the new field with the Bank code 'Bank'. 
```

```python
data['Transaction Code'] = data['Transaction Code'].str.split("-",expand = True)[0]
```

```python
data
```

```python
data.rename(columns = {'Transaction Code':'Bank'},inplace = True)
```

```python
data
```

```python
2. Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values. 
```

```python
values = {1:'Online',2:'In-Person'}
```

```python
data['Online or In-Person'] = data['Online or In-Person'].replace(values)
```

```python
data
```

```python
3. Change the date to be the day of the week 
```

```python
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'])
data['Transaction Date'] = data['Transaction Date'].dt.day_name()
```

```python
data
```

```python
4. Different levels of detail are required in the outputs. You will need to sum up the values of the transactions in three ways (help):
   (1) Total Values of Transactions by each bank
   (2) Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)
   (3) Total Values by Bank and Customer Code
```

```python
4.1.Total Values of Transactions by each bank
```

```python
banks_trans = data.groupby('Bank')['Value'].sum()
```

```python
banks_trans.reset_index(name = 'values')
```

```python
banks_trans.to_csv("Banks_Totals.csv")
```

```python
4.2.Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)
```

```python
data_trans = data.groupby(['Bank','Online or In-Person','Transaction Date'])['Value'].sum()
```

```python
data_trans.reset_index(name = "values")
```

```python
data_trans.to_csv("Bank totals by_transaction date_trans type.csv")
```

```python
4.3.Total Values by Bank and Customer Code
```

```python
data_tot = data.groupby(['Bank','Customer Code'])['Value'].sum()
```

```python
data_tot.reset_index(name = "values")
```

```python
data_tot.to_csv("Bank totals by_customer code.csv")
```

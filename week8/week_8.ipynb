```python
import os
import re
import pandas as pd
import numpy as np
```

```python
cwd = os.getcwd()
print(cwd)
```

```python
files = os.listdir(cwd)
print(files)
```

```python
df = pd.DataFrame()
```

```python
for file in files:
    if file.startswith("MOCK_DATA"):
        print(file)
        data = pd.read_csv(file)
        df = pd.concat([data.assign(month = file),df])
```

```python
df
```

```python
Create a 'file date' using the month found in the file name
  - The Null value should be replaced as 1
```

```python
df['month'] = df['month'].str.extract('(\d+)')
```

```python
df.head(5)
```

```python
df['file date'] = '01/'+ df['month'].fillna(1).astype(str)+'/2023'
```

```python
df
```

```python
Clean the Market Cap value to ensure it is the true value as 'Market Capitalisation'
   -Remove any rows with 'n/a'
```

```python
df.shape
```

```python
df['Market Cap'].isnull().sum()
```

```python
df.dropna(subset = 'Market Cap',inplace = True)
```

```python
df.shape
```

```python
Categorise the Purchase Price into groupings
   -0 to 24,999.99 as 'Low'
   -25,000 to 49,999.99 as 'Medium'
   -50,000 to 74,999.99 as 'High'
   -75,000 to 100,000 as 'Very High'
```

```python
df.info()
```

```python
df['Purchase Price'] = df['Purchase Price'].str.replace('$','')
```

```python
df['Purchase Price Category'] = pd.cut(df['Purchase Price'].astype(float),
                                      right = False,
                                      include_lowest = True,
                                      bins = [0,25_000,50_000,75_000,100_000],
                                      labels = ['Low','Medium','High','Very High'])
```

```python
df
```

```python
Categorise the Market Cap into groupings
  - Below $100M as 'Small'
  - Between $100M and below $1B as 'Medium'
  - Between $1B and below $100B as 'Large' 
  - $100B and above as 'Huge'
```

```python
df['Market Cap'] = df['Market Cap'].str.replace('$','')
```

```python
def convert_market_cap(convert_market):
    if convert_market.endswith('M'):
        return float(convert_market[:-1]) * 1e6
    if convert_market.endswith('B'):
        return float(convert_market[:-1]) * 1e9
```

```python
df['Market Cap'] = df['Market Cap'].apply(convert_market_cap)
```

```python
df['Market Cap Category'] = pd.cut(df['Market Cap'].astype(float),
                                  right = False,
                                  include_lowest = True,
                                  bins = [0,100_000_000,1_000_000_000,100_000_000_000, 999_999_999_999_999],
                                  labels = ['Small','Medium','Large','Huge'])
```

```python
df
```

```python
Rank the highest 5 purchases per combination of:
    - file date, Purchase Price Categorisation and Market Capitalisation Categorisation.
```

```python
df['Rank'] = df.groupby(['file date','Purchase Price Category','Market Cap Category'])['Purchase Price'].rank(ascending = False)
```

```python
df
```

```python
final = df[df['Rank'] <= 5]
```

```python
final
```

```python
cols = ['Market Cap Category','Purchase Price Category','File Date',
        'Ticker','Sector','Market','Stock Name','Market Cap','Purchase Price','Rank']
```

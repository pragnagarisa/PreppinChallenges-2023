```python
data_1 = pd.read_csv("Transactions.csv")
```

```python
1. In the Transactions table, there is a Sort Code field which contains dashes. We need to remove these 
   so just have a 6 digit string
```

```python
data_1['Sort Code'] = data_1['Sort Code'].str.replace("-","")
```

```python
data_1
```

```python
2. Use the SWIFT Bank Code lookup table to bring in additional information about the SWIFT code and 
   Check Digits of the receiving bank account 
```

```python
data_2 = pd.read_csv("Swift Codes.csv")
```

```python
data_2
```

```python
data_2.shape
```

```python
data_2.info()
```

```python
final_data = data_1.merge(data_2,how='inner',on='Bank')
```

```python
final_data
```

```python
3. Add a field for the Country Code 
   Hint: all these transactions take place in the UK so the Country Code should be GB
```

```python
final_data['Country Code'] = 'GB'
```

```python
final_data
```

```python
4. Create the IBAN as above 
   Hint: watch out for trying to combine sting fields with numeric fields - check data types
```

```python
final_data.info()
```

```python
final_data['IBAN'] = final_data['Country Code'] + final_data['Check Digits'] + final_data['SWIFT code'] + final_data['Sort Code'] + final_data['Account Number'].astype(str)
```

```python
final_data
```

```python
output = final_data[["Transaction ID","IBAN"]]
```

```python
output
```

```python
output.to_csv("IBAN_Output.csv")
```
